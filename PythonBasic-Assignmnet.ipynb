1. What is Python, and why is it popular?
   - Python is a high level langauge and easy to learn.
   - Python is used in fields like Data Science and Automation.
   - Its popularity comes from simple syntax, lot of librarires and flexibility across many platforms and applications.

2. What is an interpreter in Python?
   - Directly reads and executes Python code line by line, providing immediate results and making testing easy.
   - Converts source code into intermediate bytecode, which is then run by the Python Virtual Machine (PVM) to produce output.

3. What are pre-defined keywords in Python?
   - Pre-defined keywords in Python are special reserved words that have specific meanings and purposes in the language, and cannot be used as variable name.
   - eg: if, else, str, int, float, True, False, in, not, and, or, None.

4. Can keywords be used as variable names?
   - No keywords cannot be used as variable names since keywords in python are special reserved words.

5. What is mutability in Python?
   - Mutability in Python refers to whether an object value can be changed after it is created.
   - eg: list

6. Why are lists mutable, but tuples are immutable?
   - Lists are mutable because they are designed to store collections of items that may need to change—elements can be added, removed, or modified after the list is created.
   - Tuples are immutable to ensure that their contents cannot change, which provides better performance.

7. What is the difference between “==” and “is” operators in Python?
   - == checks equality of values: It returns True if the objects being compared have the same value, even if they are different objects in memory.
   - is checks identity: It returns True only if both variables point to the exact same object in memory, i.e., they have the same identity.

8. What are logical operators in Python?
   - Logical operators in Python are used to combine conditional statements and evaluate expressions as True or False.
   - and: returns True if both operands are true.
   - or: returns True if atleat one operand is true.
   - not:  Returns the opposite of the operand Boolean value.

9. What is type casting in Python?
   - Type casting in Python is the process of converting one data type into another, either automatically or explicitly by the user.
   - Python provides built-in functions like int(), float(), and str() to explicitly convert values from one type to another.
   - There are two types: Implicity and Explicity

10. What is the difference between implicit and explicit type casting?
    - Implicit type casting happens automatically by Python when converting data types during operations, without the programmer's intervention. For example, when an integer is combined with a float, Python automatically converts the integer to float to avoid data loss.
    - Explicit type casting requires the programmer to manually convert data from one type to another using built-in functions like int(), float(), or str(). This gives the programmer control over the conversion and helps avoid errors or loss of precision.

11. What is the purpose of conditional statements in Python?
    - The purpose of conditional statements in Python is to allow the program to make decisions and execute certain blocks of code based on whether specific conditions are true or false.
    - With conditionals like if, else, and elif, Python can selectively run code, such as checking if a number is even or odd or assigning grades based on scores.

12. How does the elif statement work?
    - The elif statement in Python stands for "else if" and is used within conditional blocks to check multiple conditions sequentially.
    - If the preceding if or elif condition is False, Python evaluates the next elif condition.
    - When an elif condition evaluates to True, its associated block of code executes, and the rest of the conditions are skipped.

13. What is the difference between for and while loops?
    - For loop is used when the number of iterations is known or finite. It iterates over a sequence like a list or range, executing the loop body once for each item.
    - While loop is used when the number of iterations is unknown and depends on a condition being True. It continues executing the loop body as long as the condition remains True and stops once it becomes False.

14. Describe a scenario where a while loop is more suitable than a for loop?
    - A while loop is more suitable than a for loop in scenarios where the number of iterations is unknown and depends on a condition that changes dynamically during execution.

```python
#14. Write a program to reverse a string using a while loop.
text = input()
reverse_text = ""
length = len(text)
while length > 0:
  reverse_text += text[length - 1]
  length -= 1
print(reverse_text)
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input())
fact=1
i=1
while i <= num:
  fact *= i
  i += 1
print(fact)
```
